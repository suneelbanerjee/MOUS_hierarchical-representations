SUBTLEX Lexical database cleaning



```python
import re
# Load the dataset
df = subtlex

# Ensure 'Word' is a string and handle NaN values
df['Word'] = df['Word'].astype(str).fillna('').str.strip()

# Initialize the 'Status' column to 'keep' by default
df['Status'] = 'keep'

# Ensure 'CDcount' and 'FREQcount' are numeric
df['CDcount'] = pd.to_numeric(df['CDcount'], errors='coerce')
df['FREQcount'] = pd.to_numeric(df['FREQcount'], errors='coerce')

# Remove hyphens for the length check
df['Word_no_hyphen'] = df['Word'].str.replace('-', '', regex=False)

# Filter 1: Valid word pattern
valid_word_pattern = r"^[A-Za-zÀ-ÿ]+('s)?$"
mask_valid_words = df['Word'].str.match(valid_word_pattern, na=False)

# Filter 2: Word length threshold
max_word_length = 25
mask_length = df['Word_no_hyphen'].str.len() <= max_word_length

# Filter 3: Exclude entries that are solely punctuation
mask_punctuation = ~df['Word'].str.match(r'^[^\w\s]+$', na=False)

# Filter 4: Handle 'SPEC' POS
mask_spec = df['dominant.pos'] == 'SPEC'
mask_spec_alphabetic = df['Word'].str.replace('-', '', regex=False).str.isalpha()
mask_spec_keep = ~(mask_spec & ~mask_spec_alphabetic)

# Additional Filter 1: Exclude words starting with a single letter and apostrophe
mask_no_single_letter_apostrophe = ~df['Word'].str.match(r"^[A-Za-zÀ-ÿ]'[A-Za-zÀ-ÿ]+", na=False)

# Additional Filter 2: Exclude words composed of repeated characters
def is_repeated_char(word):
    return len(set(word.lower())) == 1
mask_no_repeated_chars = ~df['Word'].apply(is_repeated_char)

# **Corrected Filter**: Flag words containing triple letters
# Use flags=re.IGNORECASE in str.contains
triple_letter_pattern = r'([a-zA-ZÀ-ÿ])\1\1'
mask_triple_letters = ~df['Word'].str.contains(triple_letter_pattern, flags=re.IGNORECASE, na=False)

# Combine all masks
mask_keep = (
    mask_valid_words & mask_length & mask_punctuation &
    mask_spec_keep & mask_no_single_letter_apostrophe &
    mask_no_repeated_chars & mask_triple_letters
)

# Update the 'Status' column
df.loc[~mask_keep, 'Status'] = 'drop'

# Remove temporary columns
df.drop(['Word_no_hyphen'], axis=1, inplace=True)

# Display counts of 'keep' and 'drop' entries
print(df['Status'].value_counts())

# Display the words marked as 'drop'
dropped_words = df[df['Status'] == 'drop']
print("\nWords to Drop:")
print(dropped_words[['Word', 'Status']].head())

# Display a sample of the DataFrame
print("\nSample of the DataFrame:")
print(df.head())
```

```python

# ------------------
# 1) Data Loading
# ------------------

# Example: we assume your final dataframe is already in 'df' after reading subtlex.
# If you already have 'df' prepared, skip the loading step and proceed to the filter masks.

# df = pd.read_csv('subtlex-nl.csv', delimiter='\t', encoding='utf-8')
# or
# df = subtlex  # if it's already in a variable named 'subtlex'

# ------------------
# 2) Prepare Data
# ------------------

# Convert 'Word' to string, handle NaN, strip whitespace
df['Word'] = df['Word'].astype(str).fillna('').str.strip()

# Ensure numeric columns
df['CDcount'] = pd.to_numeric(df['CDcount'], errors='coerce')
df['FREQcount'] = pd.to_numeric(df['FREQcount'], errors='coerce')

# Temporary column for length check (remove hyphens)
df['Word_no_hyphen'] = df['Word'].str.replace('-', '', regex=False)

# ------------------
# 3) Define Masks for Each Criterion
# ------------------

# Criterion 1: Valid word pattern
#   Matches words composed of Dutch letters (including diacritics), 
#   optionally ending in "'s".
valid_word_pattern = r"^[A-Za-zÀ-ÿ]+('s)?$"
mask_valid_words = df['Word'].str.match(valid_word_pattern, na=False)

# Criterion 2: Word length threshold
max_word_length = 25
mask_length = df['Word_no_hyphen'].str.len() <= max_word_length

# Criterion 3: Exclude entries that are solely punctuation
mask_punctuation = ~df['Word'].str.match(r'^[^\w\s]+$', na=False)

# Criterion 4: Handle 'SPEC' POS
#   Keep 'SPEC' words only if they're alphabetic (after removing hyphens).
mask_spec = df['dominant.pos'] == 'SPEC'
mask_spec_alphabetic = df['Word'].str.replace('-', '', regex=False).str.isalpha()
mask_spec_keep = ~(mask_spec & ~mask_spec_alphabetic)

# Criterion 5: Exclude words starting with a single letter and apostrophe
pattern_single_letter_apostrophe = r"^[A-Za-zÀ-ÿ]'[A-Za-zÀ-ÿ]+"
mask_no_single_letter_apostrophe = ~df['Word'].str.match(pattern_single_letter_apostrophe, na=False)

# Criterion 6: Exclude words composed entirely of repeated characters
def is_repeated_char(word):
    return len(set(word.lower())) == 1
mask_no_repeated_chars = ~df['Word'].apply(is_repeated_char)

# Criterion 7: Exclude words containing triple consecutive letters
triple_letter_pattern = r'([a-zA-ZÀ-ÿ])\1\1'
mask_triple_letters = ~df['Word'].str.contains(triple_letter_pattern, flags=re.IGNORECASE, na=False)

# ------------------
# 4) Count Failures for Each Criterion Separately
# ------------------

fail_valid_words = ~mask_valid_words
fail_length = ~mask_length
fail_punctuation = ~mask_punctuation
fail_spec = ~mask_spec_keep
fail_single_letter_apostrophe = ~mask_no_single_letter_apostrophe
fail_repeated_chars = ~mask_no_repeated_chars
fail_triple_letters = ~mask_triple_letters

print("Number of entries failing each criterion (independently):\n")
print(f"1) Invalid word pattern          : {fail_valid_words.sum()}")
print(f"2) Exceeding length threshold    : {fail_length.sum()}")
print(f"3) Solely punctuation            : {fail_punctuation.sum()}")
print(f"4) SPEC with non-alphabetic chars: {fail_spec.sum()}")
print(f"5) Single-letter + apostrophe    : {fail_single_letter_apostrophe.sum()}")
print(f"6) Entirely repeated characters  : {fail_repeated_chars.sum()}")
print(f"7) Triple consecutive letters    : {fail_triple_letters.sum()}")


# ------------------
# 5) (Optional) Combine All Masks for Final Keep/Drop
# ------------------

mask_keep = (
    mask_valid_words & mask_length & mask_punctuation & 
    mask_spec_keep & mask_no_single_letter_apostrophe &
    mask_no_repeated_chars & mask_triple_letters
)

df.loc[~mask_keep, 'Status'] = 'drop'
df.loc[mask_keep, 'Status'] = 'keep'

# Clean up the temporary column
df.drop(['Word_no_hyphen'], axis=1, inplace=True)

print("\n--- Final status counts ---")
print(df['Status'].value_counts())
```
