This script assumes that

- 2) `calculate_syllable_frequences.m` (`calculate_syllable_frequencies_loop.m`) has been run. Each auditory subject should have in their `source` folder:

- `sub-A20XX_transcription.csv`
- `sub-A20XX_word_frequencies.csv` (optional)
- `sub-A20XX_transcription_syllabes_raw.csv`
- `sub-A20XX_transcription_syllables_processed.csv`


To create a 'zero-padded' syllable frequency table, this script will merge the raw .csv (which contains onsets of all words) and the processed csv (which contains all found syllable frequency values). 

```python
import pandas as pd
from pathlib import Path

def merge_raw_processed(raw_file, processed_file):
    raw = pd.read_csv(raw_file)
    processed = pd.read_csv(processed_file)
    
    # Perform the merge and avoid duplicate column names using suffixes
    merged = pd.merge(raw, processed, on='AlignOnset', how='outer', suffixes=('', '_processed'))
    
    # If you want to drop any duplicate columns, you can identify and remove them here
    # For example, remove columns from 'processed' if their original counterpart exists
    for col in merged.columns:
        if '_processed' in col and col.replace('_processed', '') in merged.columns:
            merged.drop(col, axis=1, inplace=True)
    
    # Convert NaNs in specific columns (adjust based on actual column names)
    # For example:
    merged['Log10Minimum_Syllable_Frequency'] = np.log10(merged['Minimum_Syllable_Frequency'])
    merged['Minimum_Syllable_Frequency'].fillna(0, inplace=True)
    merged['Maximum_Syllable_Frequency'].fillna(0, inplace=True)
    merged['Mean_Syllable_Frequency'].fillna(0, inplace=True)

    # Write the merged output to a new CSV file
    output_file = Path(raw_file).parent / (Path(raw_file).stem.replace('raw', 'merged') + '.csv')
    merged.to_csv(output_file, index=False)
    return merged
```

```python
#run across all subjects
source =Path('/media/neel/MOUS/MOUS/MOUS/SynologyDrive/source')
for subject in source.iterdir():
    if subject.is_dir() and subject.name.startswith('sub-A'):
        func = subject / 'func'
        if func.is_dir():
            raw_file = func / f'{subject.name}_transcription_syllabes_raw.csv'
            processed_file = func / f'{subject.name}_transcription_syllables_processed.csv'
            if raw_file.exists() and processed_file.exists():
                merge_raw_processed(raw_file, processed_file)
print('Done')
```

```python
# Insert logarithmic values for ease of analysis later
source = Path('/media/neel/MOUS/MOUS/MOUS/SynologyDrive/source')
for subject in source.iterdir():
    if subject.is_dir() and subject.name.startswith('sub-A'):
        func = subject / 'func'
        if func.is_dir():
            merged_file = func / f'{subject.name}_transcription_syllabes_merged.csv'
            if merged_file.exists():
                merged = pd.read_csv(merged_file)
                # Recalculate log values for rows where Minimum_Syllable_Frequency > 0
                mask = merged['Minimum_Syllable_Frequency'] > 0
                merged.loc[mask, 'Log10Minimum_Syllable_Frequency'] = np.log10(merged.loc[mask, 'Minimum_Syllable_Frequency'])
                # Fill NaN values with 0
                merged['Log10Minimum_Syllable_Frequency'].fillna(0, inplace=True)
                # Save the updated DataFrame back to the CSV file
                merged.to_csv(merged_file, index=False)
                print(f"Processed {merged_file}")
            else:
                print(f"File not found: {merged_file}")
```
