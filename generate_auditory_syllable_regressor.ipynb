```python
import pandas as pd
from pathlib import Path
```

```python
syllable_table = pd.read_csv('/home/neel/Desktop/MOUS_hierarchical-representations/guslatho_MOUS_word_syllable_frequencies.csv',index_col=0)
syllable_table
```

```python
transcription_example = pd.read_csv('/media/neel/MOUS/MOUS/MOUS/SynologyDrive/source/sub-A2002/func/sub-A2002_transcription.csv')
transcription_example
```

```python
source = Path('/media/neel/MOUS/MOUS/MOUS/SynologyDrive/source')

for subject in source.rglob('sub-A*'):
    if subject.is_dir():
        # Read the transcription file
        transcription = pd.read_csv(str(subject / 'func' / f'{subject.name}_transcription.csv'))
        
        # Rename the column for consistency
        transcription.rename(columns={'Transcription': 'Word'}, inplace=True)
        transcription['Word'] = transcription['Word'].apply(lambda x: str(x).lower())
        # Merge transcription with syllable_table on the Word column
        transcription = transcription.merge(
            syllable_table[['Word', 'Min_Log10_FreqCount', 'Min_Syll_Zipf','Max_Log10_FreqCount', 'Mean_Log10_FreqCount']],
            on='Word',
            how='left'
        )
        
        # Rename columns for clarity
        transcription.rename(
            columns={
                'Min_Log10_FreqCount': 'MinLog10FreqCount',
                'Min_Syll_Zipf': 'MinSyllZipf'
            },
            inplace=True
        )
        outpath = Path(subject / 'func' / 'guslatho_syllabified.csv')
        transcription.to_csv(str(outpath))
```
