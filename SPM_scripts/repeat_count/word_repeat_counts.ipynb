```python
import pandas as pd
import csv
from pathlib import Path
import numpy as np
```

```python


source = Path('/media/neel/MOUS1/MOUS/MOUS/SynologyDrive/source')

for subject in source.iterdir():
    if subject.name.startswith('sub-A'):
        transcription = subject / 'func' / f'{subject.name}_word_frequencies.csv'
        if not transcription.exists():
            continue
        df = pd.read_csv(transcription)
# Assumes column name is exactly 'Transcription'
        df['prior_count'] = df.groupby('Transcription').cumcount()
        out_path = transcription.parent / f'{subject.name}_repeat_counts.csv'
        df.to_csv(out_path, index=False)
```
