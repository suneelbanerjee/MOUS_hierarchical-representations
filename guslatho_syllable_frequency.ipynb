```python
syllabifications = list(subtlex['Syllabification'])
syllables = [syllable for word in syllabifications if isinstance(word,str) for syllable in word.split('-')]
```

```python
Syllable_Frequency = pd.DataFrame(columns=['Syllable','Count'])
Syllable_Frequency['Syllable'] = np.unique(syllables)
Syllable_Frequency['Count'] = 0
Syllable_Frequency
```

```python
Syllable_Frequency['Count'] = Syllable_Frequency['Syllable'].map(subtlex.assign(Syllabification=subtlex['Syllabification'].str.split('-')).explode('Syllabification').groupby('Syllabification')['FREQcount'].sum()).fillna(0)
```

```python
Syllable_Frequency
```

```python
#subtlex_unaltered = pd.read_excel('SUBTLEX-NL with pos and Zipf.xlsx')

subtlex['n_syllables'] = 0
subtlex['n_syllables'] = subtlex['Syllabification'].apply(
    lambda word: len(word.split('-')) if isinstance(word, str) else 0
)
    
```

```python
dot_product = (subtlex['FREQcount'] * subtlex['n_syllables']).sum()
dot_product = dot_product / 1000000 #of syllable-sized tokens in the (filtered) SUBTLEX-NL corpus
```

```python
dot_product
```

```python
Syllable_Frequency['Syllable Zipf'] = np.log10((Syllable_Frequency['Count'] +1)/ dot_product)
minimum = min(Syllable_Frequency['Syllable Zipf'])
Syllable_Frequency
```

```python
# Syllable_Frequency['Syllable Zipf'] = Syllable_Frequency['Syllable Zipf'] + abs(minimum) + 3
minimum
```

```python
Syllable_Frequency['Syllable Zipf'] = Syllable_Frequency['Syllable Zipf'] + 3 #Following the zipf transformation for words...
```

```python
Syllable_Frequency
```

```python
Syllable_Frequency['Log10_Count'] = np.log10(Syllable_Frequency['Count']+1)
Syllable_Frequency
```

```python
#concatenate back to SUBTLEX
subtlex['Log10_Syllable_FREQcounts'] = [() for _ in range(len(subtlex))]
subtlex['Syllable_Zipfs'] = [() for _ in range(len(subtlex))]
subtlex
```

```python
# Create a mapping from syllable to its frequency and zipf value for fast lookups
syllable_map_log_count = Syllable_Frequency.set_index('Syllable')['Log10_Count']
syllable_map_zipf = Syllable_Frequency.set_index('Syllable')['Syllable Zipf']

# Define a function to apply to each row
def get_syllable_stats(syllabification):
    if not isinstance(syllabification, str):
        return (), ()
    syllables = syllabification.split('-')
    log_counts = tuple(syllable_map_log_count.get(s) for s in syllables)
    zipfs = tuple(syllable_map_zipf.get(s) for s in syllables)
    return log_counts, zipfs

# Apply the function to the 'Syllabification' column
stats = subtlex['Syllabification'].apply(get_syllable_stats)

# Assign the results back to the DataFrame
subtlex['Log10_Syllable_FREQcounts'], subtlex['Syllable_Zipfs'] = zip(*stats)
```

```python
subtlex
```

```python
subtlex_essential = subtlex[['Word','n_syllables','Log10_Syllable_FREQcounts','Syllable_Zipfs']]
subtlex_essential = subtlex_essential.drop_duplicates()
```

```python
# Handle empty tuples in 'Log10_Syllable_FREQcounts'
subtlex_essential['Min_Log10_FreqCount'] = subtlex_essential['Log10_Syllable_FREQcounts'].apply(
    lambda x: min(x) if x else None
)

subtlex_essential['Max_Log10_FreqCount'] = subtlex_essential['Log10_Syllable_FREQcounts'].apply(
    lambda x: max(x) if x else None
)
subtlex_essential['Mean_Log10_FreqCount'] = subtlex_essential['Log10_Syllable_FREQcounts'].apply(
    lambda x: np.mean(x) if x else None
)

# Handle empty tuples in 'Syllable_Zipfs'
subtlex_essential['Min_Syll_Zipf'] = subtlex_essential['Syllable_Zipfs'].apply(
    lambda x: min(x) if x else None
)

subtlex_essential
```
